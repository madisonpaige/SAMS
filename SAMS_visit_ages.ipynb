```python
SAMS_visit_ages = pd.DataFrame()
```

```python
data['intake_date_w1'] = pd.to_datetime(data['intake_date_w1'])
data['cni_w1'] = pd.to_datetime(data['cni_w1'])
data['lp_w1'] = pd.to_datetime(data['lp_w1'])
data['intake_date_w2'] = pd.to_datetime(data['intake_date_w2'])
data['lp_w2'] = pd.to_datetime(data['lp_w2'])
data['pet_w2'] = pd.to_datetime(data['pet_w2'])
data['dob'] = pd.to_datetime(data['dob'])
```

```python
print(data)
```

```python
SAMS_visit_ages['pidn'] = data['pidn']
SAMS_visit_ages['intake_w1_age'] = (data['intake_date_w1'] - data['dob'])
SAMS_visit_ages['cni_w1_age'] = data['cni_w1'] - data['dob']
SAMS_visit_ages['lp_w1_age'] = data['lp_w1'] - data['dob']
SAMS_visit_ages['intake_w2_age'] = data['intake_date_w2'] - data['dob']
SAMS_visit_ages['lp_w2_age'] = data['lp_w2'] - data['dob']
SAMS_visit_ages['pet_w2_age'] = data['pet_w2'] - data['dob']
```

```python
row = 0 

for n in pd.unique(SAMS_visit_ages['pidn']):
    temp = (str(SAMS_visit_ages.loc[row, 'intake_w1_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'intake_w1_age'] = (pd.to_numeric(temp[0]))/365.25
        
    temp = (str(SAMS_visit_ages.loc[row, 'cni_w1_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'cni_w1_age'] = pd.to_numeric(temp[0])/365.25
        
    temp = (str(SAMS_visit_ages.loc[row, 'lp_w1_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'lp_w1_age'] = (pd.to_numeric(temp[0]))/365.25
        
    temp = (str(SAMS_visit_ages.loc[row, 'intake_w2_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'intake_w2_age'] = pd.to_numeric(temp[0])/365.25
        
    temp = (str(SAMS_visit_ages.loc[row, 'lp_w2_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'lp_w2_age'] = (pd.to_numeric(temp[0]))/365.25
        
    temp = (str(SAMS_visit_ages.loc[row, 'pet_w2_age'])).split(' ')
    if temp[0] != 'NaT':
        SAMS_visit_ages.loc[row, 'pet_w2_age'] = pd.to_numeric(temp[0])/365.25
        
    row+=1
```
