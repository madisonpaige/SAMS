```python
#set up
import pandas as pd
import os
import glob
import datetime as dt
from dateutil.relativedelta import relativedelta
```

```python
#import csv with data
path = '/Users/madisonhunt/Desktop/Scheduling.csv'
sched_data = pd.read_csv(path)
sched_data.head()
```

```python
#create a data frame of consent dates formatted as dates
date_list = pd.DataFrame()

row=0    
for f in sched_data['record_num']:
    f = date_list.loc[row, 'record_num'] = str(f)
    row+=1

row = 0
for l in sched_data['first_name']:
    date_list.loc[row, 'first_name'] = l
    row+=1
row = 0
for k in sched_data['last_name']:
    date_list.loc[row, 'last_name'] = k
    row+=1
    
row = 0
for n in sched_data['consent_w1']:
    n = pd.to_datetime(n)
    date_list.loc[row, 'cons_date'] = n
    row+=1
    
date_list.dtypes
```

```python
# calculate beginning of contact window
row = 0
for m in date_list['cons_date']:
    date_list.loc[row, 'contact_date'] = (m+ relativedelta(months=+39))
    row+=1
    
                     
```

```python
   # calculate eligibility date
row = 0
for m in date_list['cons_date']:
    date_list.loc[row, 'eligib_date'] = (m+ relativedelta(months=+42))
    row+=1
            
```

```python
# calculate end of eligibility date
row = 0
for m in date_list['cons_date']:
    date_list.loc[row, 'end_eligib_date'] = (m+ relativedelta(months=+45))
    row+=1
    
print(date_list)
```

```python
date_list.to_csv('/Users/madisonhunt/Desktop/w2_sched_info.csv')
```
